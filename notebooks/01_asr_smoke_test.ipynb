```python
import numpy as np
import sounddevice as sd
import whisper
import re
```

```python
DURATION = 7
SAMPLE_RATE = 16000

print("Recording... speak Arabic clearly الآن")
audio = sd.rec(int(DURATION * SAMPLE_RATE),
               samplerate=SAMPLE_RATE,
               channels=1,
               dtype="float32")
sd.wait()
print("Recording finished.")

audio = audio.flatten()
print("Audio mean abs:", float(np.abs(audio).mean()))
```

```python
model = whisper.load_model("small")  # better Arabic than "base"

result = model.transcribe(
    audio,
    language="en",
    task="transcribe",
    fp16=False,
    temperature=0.0,
    no_speech_threshold=0.6,
    logprob_threshold=-1.0,
    condition_on_previous_text=False
)

print("Raw Recognized Text:")
print(result["text"])
```

```python
raw_text = result["text"]

clean_text = re.sub(r"[^\u0600-\u06FF0-9\s\?\!\.\،\؛\:]", "", raw_text)
clean_text = re.sub(r"\s+", " ", clean_text).strip()

print("Clean Arabic Text:")
print(clean_text)
```

```python
print("Detected language:", result.get("language"))
print("Segments:", len(result.get("segments", [])))
if result.get("segments"):
    print("First segment:", result["segments"][0]["text"])
```
